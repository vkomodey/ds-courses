1. Достать 100 картинок дронов и 100 картинок не дронов размером 640x480
2. Расширить картинки различными преобразованиями

Все картинки находятся в gdrive - https://drive.google.com/file/d/1sytBukpliftknIKYUwlXx31LQAm5KfMB/view?usp=sharing
Распаковать архив необходимо в корень задания - /dataset/{positive,negative}

```python
from skimage import io
import warnings
warnings.filterwarnings("ignore")
```

```python
positive_dataset = io.imread_collection('./dataset/positive/*');
negative_dataset = io.imread_collection('./dataset/negative/*');
```

```python
from skimage.transform import rotate
from scipy import ndimage
from skimage.util import random_noise

def augment_images(images):
    augmented_images = []

    augmented_images += list(map(lambda image: random_noise(image), images)) # add noize
    augmented_images += list(map(lambda image: ndimage.uniform_filter(image, size=(11, 11, 1)), images)) # blur
    augmented_images += list(map(lambda image: rotate(image, 45), images)) # 45 degree rotate
    augmented_images += list(map(lambda image: image[:, ::-1], images)) # flip
    
    return augmented_images
```

```python
augmented_positive = augment_images(positive_dataset)
augmented_negative = augment_images(negative_dataset)
```

```python
for i, image in enumerate(augmented_negative):
    io.imsave('./dataset/negative/augmented_{}.jpg'.format(i), image);
```

```python
for i, image in enumerate(augmented_positive):
    io.imsave('dataset/augmented_{}.jpg'.format(i), image);
```
